"Write a function `is_prime(n)` that accepts an argument n and returns boolean `True` if n is a prime number and boolean `False` if n is not prime. For example, calling `is_prime(11)` should return `True` and `is_prime(12)` should return `False`." via PPD 534 Assignment Instructions

```python
def is_prime(n):
  if n <= 1:
    return False
  else:
    is_prime = True #used help from Google
    for egg in range(2,(n-1)):
      if n % egg == 0:
        return False
        break #used help from Google
    return True
```

#2ii.
"Write a function `print_primes(n)` that accepts an argument n and prints all the prime numbers between 0 and n (inclusive). This function should call the `is_prime` function you developed earlier. For example, calling `print_primes(13)` should print out: `2 3 5 7 11 13`" via PPD 534 Assignment Instructions

```python
def print_primes(n):
  if n <= 1:
    return []
  else:
    primes_list = []
    for beep in range(2,n):
      if is_prime(beep):
        primes_list.append(beep)
    return primes_list
```

```python
print_primes(3)
```

```python
print_primes(25)
```

# 2iii.
"Write a function `count_primes(n)` that accepts an argument `n` and returns an integer count of how many prime numbers there are between 0 and n (inclusive). This function should also call the `is_prime` function you developed earlier. For example, calling `count_primes(19)` should return 8, because there are eight prime numbers between 0 and 19."
via PPD 534 Assignment Instructions

```python
def count_primes(n):
  prime_count = 0             # create a variable called prime_count and make it start at 0
  for x in range (0, n+1):    # for loop; set the function to repeat for every value between a range
    if is_prime(x) == True:
      prime_count += 1        # if the number is prime, add 1 to the count_primes variable
  return prime_count
```

```python
count_primes(19)
```

```python
count_primes(100)
```

# 2iv.
"Write a function `first_primes(n)` that accepts an argument n and returns a Python list containing the first n prime numbers. (Hint: use a while loop and append each prime you find to a list.) This function should also call the `is_prime` function you developed earlier. For example, calling `first_primes(5)` should return the first five prime numbers as a list: `[2, 3, 5, 7, 11]`"
via PPD 534 Assignment Instructions

```python
def first_primes(n):
  first_primes_list = []              # make a list to append to later
  x = 0                               # start the count at 0
  while len(first_primes_list) < n:   # the while loop will continue until it hits n
    if is_prime(x) == True:
      first_primes_list.append(x)     # add the found prime number to the list
    x += 1                            # continues the loop
  return first_primes_list
```

```python
first_primes(5)
```

```python
first_primes(20)
```
